## Indexing

```python
import pandas as pd
```

```python
df = pd.read_csv("files/world_population.csv")
```

```python
df.set_index("Country", inplace=True)
```

```python
df.head()
```

```python
df.loc['Albania']
```

```python
df.loc['Albania', ['2022 Population', "Area (km²)"]]
```

```python
df2 = df.loc[['Albania', 'Bulgaria'], ['2022 Population', "Area (km²)"]]
df2
```

```python
df2 = df.loc[:, ['2022 Population', "Area (km²)"]]
df2
```

```python
df.iloc[0:11, [2,3,5]]
```
