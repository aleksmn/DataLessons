## Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Style

print(plt.style.available)
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("files/world_population.csv")
df[50:60]
```

```python
# df[df['Country'].isin(["Russia", "China"])]
```

```python
df = df.set_index("Country")
```

```python
df = df.filter(items=["Russia", "China", "United States", "Brazil"], axis=0)
```

```python
df = df.iloc[:, 4:12]
```

```python
df
```

```python
df = df.transpose()
df
```

```python
df.index = ['2022', '2020', '2015',
       '2010', '2000', '1990',
       '1980', '1970']
```

```python
df.sort_index(inplace=True)
```

```python
df.plot()
```

```python
df.plot(kind="line", subplots=False, title="Population", figsize=(8, 5))
```

## Bar plot

```python
df = pd.read_csv("files/values.csv")
df
```

```python
df = df.set_index("date")
```

```python
df.plot(kind="bar", title="Values", xlabel="Date", figsize=(8, 5))
```

```python
df.plot(kind="bar", stacked=True, title="Values", xlabel="Date", figsize=(8, 5))
```

```python
df.plot(kind="barh")
```

## Histogram

```python
df.plot(kind="hist", bins=10, stacked=True)
```

## Scatter

```python
df.plot(kind="scatter", x="value 1", y="value 3", s=100)
```
