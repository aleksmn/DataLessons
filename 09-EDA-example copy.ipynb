# Step 2: Data Preparation

```python
df = df[['coaster_name',
     'opening_date_clean',
    'Location', 
    'Status',
    'Manufacturer',
    'year_introduced',
    'latitude',
    'longitude',
    'Type_Main',
    'speed_mph', 
    'height_ft',
    'Inversions_clean', 
    'Gforce_clean']].copy()
```

```python
df.head()
```

```python
df = df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'})
```

```python
df.head()
```

```python
df['Opening_Date'] = pd.to_datetime(df['Opening_Date'])
```

```python
# Проверка на дубли
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(subset=["Coaster_Name"])]
```

```python
df.query("Coaster_Name == 'American Dreier Looping'")
```

```python
df = df.loc[~df.duplicated(subset=["Coaster_Name"])].reset_index(drop=True).copy()

df
```
