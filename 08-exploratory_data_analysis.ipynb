### Correlation

```python

df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlGnBu")
# cmaps: rocket viridis mako magma YlOrBr vlag  icefire coolwarm YlGnBu

plt.rcParams['figure.figsize'] = 12,10

plt.show()
```

```python
df.groupby('Continent').mean(numeric_only=True)
```

```python
oceania = list(df[df["Continent"].str.contains("Oceania")]["Country"])
```

```python
def get_counties_list(continent):
    return list(df[df["Continent"].str.contains(continent)]["Country"])
```

```python
print((get_counties_list("Europe")))
```

```python
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
df2
```

```python
df2.columns
```

```python
df2 = df2.loc[:, ['2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population']]
```

```python
df2 = df2.transpose()
```

```python
df2.index = ['2022', '2020', '2015',
       '2010', '2000', '1990',
       '1980', '1970']
```

```python
df2 = df2.sort_index()
df2
```

```python
df2.plot(title="World Population")
```
