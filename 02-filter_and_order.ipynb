## Filtering

```python
import pandas as pd
```

```python
df = pd.read_csv("files/world_population.csv")
```

```python
df
```

```python
df[df["Rank"] <= 10]
```

```python
countries = ["India", "China", "Russia"]

df[df['Country'].isin(countries)]
```

```python
df[df['Country'].str.contains("Islands")]
```

```python
df[df['Country'].str.startswith("I")]
```

## Set index

```python
df2 = df.set_index("Country")
df2
```

```python
df2.filter(items=["2022 Population"], axis=1)
```

```python
df2.filter(items=["Russia", "Brazil"], axis=0)
```

```python
df2.filter(like="Islands", axis=0)
```

```python
df2.loc["Russia"]
```

```python
df3 = df2.iloc[:, [3, 4]]
```

```python
df3[df3["2022 Population"] > 10**9]
```

## Order By

```python
df[df["Rank"] <= 10].sort_values(by="Rank", ascending=False)
```

```python
df[df["Rank"] <= 20].sort_values(by=["Continent", "Country"])
```
